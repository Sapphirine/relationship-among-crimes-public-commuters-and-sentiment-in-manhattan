```python
import pandas as pd
import numpy as np
```

```python
GAP = 0.004
```

```python
df = pd.read_csv("./NYPD_Complaint_Data_Historic.csv")
df.describe()
```

```python
df.columns.values
```

```python
df = df.drop(['CMPLNT_NUM', 'CMPLNT_TO_DT', 'CMPLNT_TO_TM',
              'RPT_DT', 'KY_CD', 'OFNS_DESC', 'PD_CD', 'PD_DESC', 'CRM_ATPT_CPTD_CD',
              'LAW_CAT_CD', 'JURIS_DESC', 'BORO_NM', 'LOC_OF_OCCUR_DESC', 
              'PREM_TYP_DESC', 'PARKS_NM', 'HADEVELOPT', 'X_COORD_CD', 'Y_COORD_CD',
              'Lat_Lon'], axis=1)
```

```python
df.columns = ["Datetime", "Time", "Precinct", "Latitude", "Longitude"]
```

```python
df = df.dropna()
```

```python
df.describe()
```

```python
df["Datetime"] = df["Datetime"].astype(str)
df["Time"] = df["Time"].astype(str)
df["Precinct"] = df["Precinct"].astype(str)
df["Latitude"] = df["Latitude"].astype(str)
df["Longitude"] = df["Longitude"].astype(str)
```

```python
df.head()
```

```python
# df = df[~df.Datetime.str.contains("nan")]
# df = df[~df.Time.str.contains("nan")]
# df = df[~df.Precinct.str.contains("nan")]
# df = df[~df.Longitude.str.contains("nan")]
# df = df[~df.Latitude.str.contains("nan")]
```

```python
df["Latitude"] = pd.to_numeric(df["Latitude"])
df["Longitude"] = pd.to_numeric(df["Longitude"])
```

```python
labels_lat = ["{0}".format(i) for i in range(len(np.arange(40.494292,40.925521,GAP)))]
df['group_lat'] = pd.cut(df.Latitude, np.arange(40.494292,40.925521+GAP,GAP), labels=labels_lat)
labels_long = ["{0}".format(i) for i in range(len(np.arange(-74.254380,-73.693835,GAP)))]
df['group_long'] = pd.cut(df.Longitude, np.arange(-74.254380,-73.693835+GAP,GAP), labels=labels_long)
```

```python
datetime_arr = df["Datetime"].values
time_arr = df["Time"].values
```

```python
import datetime
day_list = []
for time in datetime_arr:
    time_obj = datetime.datetime.strptime(time, "%m/%d/%Y")
    day_list.append(time_obj.weekday())
    
hour_list = []
for time in time_arr:
    time_obj = datetime.datetime.strptime(time, "%H:%M:%S")
    hour_list.append(time_obj.hour)
```

```python
day_se = pd.Series(day_list)
hour_se = pd.Series(hour_list)
```

```python
df["day"] = day_se
df["hour"] = hour_se
df = df.drop(["Datetime", "Time"], axis=1)
```

```python
df
```

```python
df = df.dropna()
```

```python
df.describe()
```

```python
df["day"] = df["day"].astype(int)
df["hour"] = df["hour"].astype(int)
```

```python
df = df.reset_index(drop=True)
```

```python
df.describe()
```

```python
df["num"] = 1
```

```python
df_final = pd.DataFrame({"sum" : df.groupby(["group_lat","group_long","day","hour"])["num"].sum()}).reset_index()
```

```python
df_final.head()
```

```python
df_final["group_lat"] = df_final["group_lat"].astype(int)
df_final["group_long"] = df_final["group_long"].astype(int)
```

```python
def numeric_to_lat(numeric):
    result = 40.494292 + (numeric+1/2)*GAP
    return result

def numeric_to_long(numeric):
    result = -74.254380 + (numeric+1/2)*GAP
    return result
```

```python
df_final["lat"] = df_final["group_lat"].map(numeric_to_lat)
df_final["long"] = df_final["group_long"].map(numeric_to_long)
```

```python
df_return = df_final[["day","hour","sum","lat","long"]]
```

```python
df_return
```

```python
df_return.to_csv("./criminal_lat_long_%s.csv"%(str(GAP-int(GAP))[2:]),index = None)
```
